```python
import os
import pandas as pd

# Check for files in the current directory
files = os.listdir('.')
print("Files available:", files)

# If BlinkIT Grocery Data.xlsx exists, check its sheets
if "BlinkIT Grocery Data.xlsx" in files:
    xls = pd.ExcelFile("C:/Users/hp/Downloads/BlinkIT Grocery Data.xlsx")
    print("Sheets in BlinkIT Grocery Data.xlsx:", xls.sheet_names)
```

```python
df = pd.read_excel("C:/Users/hp/Downloads/BlinkIT Grocery Data.xlsx", sheet_name='BlinkIT Grocery Data')
print("Shape:", df.shape)
print("\nColumns and Data Types:")
print(df.dtypes)
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("Missing values:")
print(df.isnull().sum())

print("\nUnique counts:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
    if df[col].nunique() < 10:
        print(df[col].unique())
```

```python
# Clean Item Fat Content
df['Item Fat Content'] = df['Item Fat Content'].replace({'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'})
print("Cleaned Item Fat Content:", df['Item Fat Content'].unique())

# Core summary metrics
total_sales = df['Sales'].sum()
avg_sales = df['Sales'].mean()
total_items = df['Item Identifier'].nunique()
total_outlets = df['Outlet Identifier'].nunique()
avg_rating = df['Rating'].mean()
avg_visibility = df['Item Visibility'].mean()

print(f"Total Sales: {total_sales}")
print(f"Average Sales: {avg_sales}")
print(f"Total Items: {total_items}")
print(f"Total Outlets: {total_outlets}")
print(f"Average Rating: {avg_rating}")
print(f"Average Visibility: {avg_visibility}")
```

```python
# Sales by Year
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().reset_index().sort_values('Outlet Establishment Year')
print("Sales by Year:\n", sales_by_year)

# Sales by Item Type
sales_by_item = df.groupby('Item Type')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
print("\nSales by Item Type:\n", sales_by_item)

# Sales by Outlet Type / Location Type
sales_by_outlet_type = df.groupby('Outlet Type')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
print("\nSales by Outlet Type:\n", sales_by_outlet_type)

sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
print("\nSales by Location:\n", sales_by_location)
```

```python
outlets_by_year = df.groupby('Outlet Establishment Year')['Outlet Identifier'].nunique().reset_index()
print("Outlets established by year:\n", outlets_by_year)

# Check average sales per outlet type
outlet_perf = df.groupby('Outlet Identifier').agg(
    Year=('Outlet Establishment Year', 'first'),
    Type=('Outlet Type', 'first'),
    Location=('Outlet Location Type', 'first'),
    Size=('Outlet Size', 'first'),
    Total_Sales=('Sales', 'sum'),
    Item_Count=('Item Identifier', 'count')
).reset_index()
print("\nOutlet Performance Table:\n", outlet_perf)
```
